```python
# Task 2: Mark the success/failed launches for each site on the map

import folium
from folium.features import DivIcon

# Coordinates of NASA Johnson Space Center
nasa_coordinate = [29.559684888585615, -95.0830971930759]

# Initialize the map
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# List of launches with coordinates, site names, and outcomes
launches = [
    {"site": "CCAFS LC-40", "Lat": 28.563302, "Long": -80.577356, "outcome": "Success"},
    {"site": "CCAFS SLC-40", "Lat": 28.561857, "Long": -80.577320, "outcome": "Failed"},
    {"site": "KSC LC-39A", "Lat": 28.573255, "Long": -80.646895, "outcome": "Success"},
    {"site": "VAFB SLC-4E", "Lat": 34.632834, "Long": -120.610746, "outcome": "Failed"}
]

# Add a marker for NASA Johnson Space Center
folium.Marker(
    location=nasa_coordinate,
    popup='NASA Johnson Space Center',
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'NASA JSC',
    )
).add_to(site_map)

# Add circles and markers for each launch site based on the launch outcome
for launch in launches:
    coordinate = [launch["Lat"], launch["Long"]]
    color = 'green' if launch["outcome"] == "Success" else 'red'
    folium.Circle(
        location=coordinate,
        radius=1000,
        color=color,
        fill=True,
        fill_color=color
    ).add_child(folium.Popup(f'{launch["site"]}: {launch["outcome"]}')).add_to(site_map)

    folium.Marker(
        location=coordinate,
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % launch["site"],
        )
    ).add_to(site_map)

# Display the map
site_map
```

Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
```

```python

launch_site_lat = 28.56367
launch_site_lon = -80.57163
closest_city_lat = 28.55600  
closest_city_lon = -80.58800  

# Initialize your map
site_map = folium.Map(location=[(launch_site_lat + closest_city_lat) / 2, (launch_site_lon + closest_city_lon) / 2], zoom_start=12)

# Calculate the distance to the closest city
distance_city = calculate_distance(launch_site_lat, launch_site_lon, closest_city_lat, closest_city_lon)

# Create and add the marker for the closest city
city_marker = folium.Marker(
    location=[closest_city_lat, closest_city_lon],
    icon=DivIcon(
        icon_size=(150, 36),
        icon_anchor=(7, 20),
        html='<div style="font-size: 12px; color:#d35400;"><b>{:.2f} KM</b></div>'.format(distance_city),
    )
)
city_marker.add_to(site_map)

# Create PolyLine
coordinates = [
    [launch_site_lat, launch_site_lon],
    [closest_city_lat, closest_city_lon]
]
line_to_city = folium.PolyLine(locations=coordinates, weight=2, color='red')
site_map.add_child(line_to_city)

# Display the map
site_map
```

```python
# Define proximity thresholds (in km)
proximity_threshold = {
    'railway': 10,  
    'highway': 10,
    'coastline': 10,
    'city': 20  
}

# coordinates for analysis
closest_railway_lat = 28.57000
closest_railway_lon = -80.57500
closest_highway_lat = 28.58000
closest_highway_lon = -80.58000
closest_city_lat = 28.55600
closest_city_lon = -80.58800
coastline_lat = 28.56367
coastline_lon = -80.57163

# Calculate distances
distance_railway = calculate_distance(launch_site_lat, launch_site_lon, closest_railway_lat, closest_railway_lon)
distance_highway = calculate_distance(launch_site_lat, launch_site_lon, closest_highway_lat, closest_highway_lon)
distance_city = calculate_distance(launch_site_lat, launch_site_lon, closest_city_lat, closest_city_lon)
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Analyze proximities
def analyze_proximity(distance, threshold):
    return distance <= threshold

# Proximity results
proximity_results = {
    'railway': analyze_proximity(distance_railway, proximity_threshold['railway']),
    'highway': analyze_proximity(distance_highway, proximity_threshold['highway']),
    'coastline': analyze_proximity(distance_coastline, proximity_threshold['coastline']),
    'city': not analyze_proximity(distance_city, proximity_threshold['city'])  # Cities should be farther than the threshold
}

# Output results
print(f"Is the launch site close to railways? {'Yes' if proximity_results['railway'] else 'No'}")
print(f"Is the launch site close to highways? {'Yes' if proximity_results['highway'] else 'No'}")
print(f"Is the launch site close to the coastline? {'Yes' if proximity_results['coastline'] else 'No'}")
print(f"Does the launch site maintain a certain distance from cities? {'Yes' if proximity_results['city'] else 'No'}")
```
